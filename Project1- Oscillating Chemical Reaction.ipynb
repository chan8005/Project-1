# Oscillating Chemical Reactions


```python
import matplotlib.pylab as plt
import numpy as np

#set up for future plots

#Absolute constants
t_init= 0          # inital time
t_final= 50       # time to end data collection
N= 100000            # number of steps to take between start and end time
h= (t_final - t_init)/N  #delta t

def dx(x,y,a,b):
    """Calculates rate of change of chemical x at a given time.
    x= concentration of chemical x at given time
    y= concentration of chemical y at given time
    a= positive constant
    b= positive constant"""
    return 1-(b+1)*x + a*y*x**2

def dy(x,y,a,b):
    """Calculates rate of change of chemical y at a given time
    x= concentration of chemical x at given time
    y= concentration of chemical y at given time
    a= positive constant
    b= positive constant"""
    return b*x - a*y*x**2

def plots(a,b,x,y):
    tpoints = np.arange(t_init,t_final,h)       #array for time that goes from initial time to final time in increments of h
    xarray=[]                     #empty list for [x] at given time
    yarray=[]                     # empty list for [y] at given time

    for t in tpoints:            # for every given time...
        xarray.append(x)          # appending [x], first one is the initial [], the next one will be calculated later in loop
        yarray.append(y)            # appending the initial [y]
        newx= h* dx(x,y,a,b)         # calculates the next component of [x] by multiplying rate of change of [x] times delta t 
        newy = h* dy(x,y,a,b)        # calculates the next component of [y] by multiplying rate of change of [y] times delta t 
        x += newx   # adds calculated components to the old [x] to calculate the new [x] which will be added to array when loop restarts
        y += newy   # adds calculated components to the old [x] to calculate the new [x] which will be added to array when loop restarts
    return xarray,yarray,tpoints

```

```python
#changing initial concentraitons with fixed ab values
a= 1               # constant
b= 3

xy= [(0,0),(1,10),(10,0)]

for n in xy:
    x= n[0]
    y= n[1]
    
    print ("Initial [x] is:",x)
    print ("Initial [y] is:",y)
    xarray,yarray,tpoints = plots(a,b,x,y)

    plt.figure(figsize=(10, 10))
    plt.subplot (221)
    plt.plot(tpoints,xarray,"b.", label= "[x]")
    plt.plot(tpoints,yarray,"r.",label= "[y]")
    plt.ylabel("Concentration")
    plt.xlabel("time")
    plt.title("Concentrations of Chemical x and y vs time")

    plt.subplot (222)
    plt.plot(xarray,yarray,".")

    plt.ylabel("[y]")
    plt.xlabel("[x]")
    plt.axis('equal')
    plt.title("Concentrations of Chemical x vs chemical y")
    legend = plt.legend(loc='upper right', shadow=True)
    plt.show()
    print("--------------------------------------------------------------------------------------------------------")
```

The above cell explores how changing the initial concentration of x and y affect the oscillations of the chemical reaction.  The graphs depict the concentrations of chemical x and chemical y vs time and concentration of x vs y using Euler’s method to get the data points.  Regardless of the initial concentrations of x and y, all the graphs depict the same trends. For the graphs that depict the [x] and [y], the starting points of the initial conditions all go to the same oscillating trends. Once the concentrations start oscillating, the graphs look the same regardless of the initial concentrations. This indicates that the concentrations of x and y are not relevant in the oscillating nature of this reaction. This makes sense because we still see an oscillation of the reaction when x and y are zero to begin with. This means that these species are created throughout the mechanism and other factors in the mechanism dictate how much of it is created or degraded.
